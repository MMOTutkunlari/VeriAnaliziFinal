### Yaşadığı Bölgeye Göre İnme Belirtileri

```python
yasadigiBolge = df.Residence_type.unique()
print(yasadigiBolge)
```

```python
felcli = []
felcsiz = []

for y in yasadigiBolge:
    felcli.append(len(df[df["Residence_type"] == y and df["stroke"] == 1]))
    felcsiz.append(len(df[df["Residence_type"] == y and df["stroke"] == 0]))
```

```python
df.drop(["id"], axis=1, inplace=True)
```

```python
df["gender"]=[1  if i=="Male" else 0  for i in df["gender"]]
```

```python
df["ever_married"]=[1  if i=="Yes" else 0  for i in df["ever_married"]]
```

```python
df
```

```python
for i in range(len(df["work_type"])):
    if (df["work_type"][i] == "Private" or df["work_type"][i] == "Self-employed" or df["work_type"][i] == "Govt_job"):
        df["work_type"][i] = "1"
    else:
        df["work_type"][i] = "0"
```

```python
df
```

```python
Residence_type = pd.get_dummies(df["Residence_type"], prefix="Residence_type")
smoking_status = pd.get_dummies(df["smoking_status"], prefix="smoking_status")
```

```python
df=pd.concat([df, Residence_type], axis=1)
df=pd.concat([df, smoking_status], axis=1)
```

```python
df
```

```python
df.drop(["Residence_type", "smoking_status"], axis=1, inplace=True)
```

```python
df.dropna(subset = ["bmi"], inplace=True)
```

```python
df.info()
```

```python
df["stroke"].value_counts()
```

```python
df["work_type"] = df["work_type"].astype("uint8")
```

```python
y = df["stroke"]
x_data = df.drop(["stroke"], axis=1)
```

```python
x = ( x_data - np.min(x_data) ) / ( np.max(x_data) - np.min(x_data) ).values
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=32)
```

```python
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
nb=GaussianNB()
nb.fit(x_train,y_train)
print("Naive Bayes Skor: ",nb.score(x_test,y_test))
```

```python
tahminler=nb.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,tahminler)
import seaborn as sn
plt.figure(figsize=(8,10))
sn.heatmap(cm, annot=True, fmt=".3g")
plt.xlabel("Gerçek Değer")
plt.ylabel("Tahminler")
plt.show()
```

```python
skor_liste=[]
for i in range(1,10):
    knn=KNeighborsClassifier(n_neighbors=i)  #k değeri seçiliyor
    knn.fit(x_train,y_train)
    skor_liste.append(knn.score(x_test,y_test))
```

```python
plt.plot(range(1,10),skor_liste)
plt.xlabel("k değeri")
plt.ylabel("Skor")
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(x_train,y_train)
print(knn.score(x_test,y_test))
```

```python
tahminler1=knn.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,tahminler1)
import seaborn as sn
plt.figure(figsize=(8,10))
sn.heatmap(cm, annot=True, fmt=".3g")
plt.xlabel("Gerçek Değer")
plt.ylabel("Tahminler")
plt.show()
```

```python
y_test.value_counts()
```

```python
dt=DecisionTreeClassifier()
dt.fit(x_train, y_train)
print(dt.score(x_test,y_test))
```

```python
tahminler2=dt.predict(x_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,tahminler2)
import seaborn as sn
plt.figure(figsize=(8,10))
sn.heatmap(cm, annot=True, fmt=".3g")
plt.xlabel("Gerçek Değer")
plt.ylabel("Tahminler")
plt.show()
```

```python
modeller=["NB","KNN","DT"]
skorlar=[nb.score(x_test,y_test),knn.score(x_test,y_test),dt.score(x_test,y_test)]
```

```python
plt.scatter(modeller,skorlar)
plt.plot(modeller,skorlar, color="red")
plt.xlabel("Modeller")
plt.ylabel("Skorlar")
plt.show()
```
